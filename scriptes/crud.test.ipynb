```python
from pymongo import MongoClient
```

```python


client = MongoClient("mongodb://localhost:27017")
db = client["weather_project"]
col = db["infoclimat"]

print("\n===== TEST CRUD =====")

# CREATE
doc_id = col.insert_one({
    "station_id": "TEST123",
    "name": "Station Test",
    "location": {"latitude": 10, "longitude": 20}
}).inserted_id
print(f"CREATE -> OK id={doc_id}")

# READ
doc = col.find_one({"station_id": "TEST123"})
print("READ ->", doc)

# UPDATE
col.update_one(
    {"station_id": "TEST123"},
    {"$set": {"name": "Station Test Updated"}}
)
print("UPDATE -> OK")

# DELETE
col.delete_one({"station_id": "TEST123"})
print("DELETE -> OK")

print("\nCRUD terminé ")
```
