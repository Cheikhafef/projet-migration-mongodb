## Conversion Excel ‚Üí JSON ‚Äî Weather Underground (La Madeleine, FR)
Ce script transforme les donn√©es m√©t√©orologiques issues d‚Äôun fichier Excel Weather Underground en un fichier JSON harmonis√©, conforme au sch√©ma utilis√© dans le pipeline de traitement.

## üîß Fonctionnement g√©n√©ral
Charge toutes les feuilles du fichier Excel (sheet_name=None)

Ajoute un identifiant de station fixe : ILAMAD25

Ajoute une colonne date correspondant au nom de la feuille (ex. 2024-10-01)

Renomme les colonnes pour correspondre au sch√©ma du pipeline

Concat√®ne toutes les feuilles en un seul DataFrame

Exporte le r√©sultat final dans un fichier JSON format√© et lisible

```python
import pandas as pd

excel_file = "Weather Underground - La Madeleine, FR.xlsx"

# Charger toutes les feuilles
sheets = pd.read_excel(excel_file, sheet_name=None)

all_data = []

for sheet_name, df in sheets.items():

    # Ajouter station_id
    df["station_id"] = "ILAMAD25"

    # Ajouter la colonne date = nom de la feuille
    df["date"] = sheet_name

    # Renommer les colonnes pour correspondre √† ton pipeline
    df = df.rename(columns={
        "Time": "time",
        "Temperature": "temperature",
        "Dew Point": "dew_point",
        "Humidity": "humidity",
        "Wind": "wind_direction",
        "Speed": "wind_speed",
        "Gust": "wind_gust",
        "Pressure": "pressure",
        "Precip. Rate.": "precip_rate",
        "Precip. Accum.": "precip_accum",
        "UV": "uv",
        "Solar": "solar"
    })

    # Ajouter au tableau global
    all_data.append(df)

# Fusionner toutes les feuilles
final_df = pd.concat(all_data, ignore_index=True)

# Export JSON
final_df.to_json(
    "la_madeleine.json",
    orient="records",
    indent=4,
    force_ascii=False
)

print("Conversion termin√©e ‚Üí la_madeleine.json")
```
