## ðŸ“„ Conversion Excel â†’ JSON (Weather Underground â€“ Ichtegem)

Ce script charge lâ€™ensemble des feuilles dâ€™un fichier Excel Weather Underground, harmonise les colonnes pour correspondre au pipeline de traitement, ajoute les mÃ©tadonnÃ©es nÃ©cessaires, puis gÃ©nÃ¨re un fichier JSON final prÃªt pour ingestion.

### ðŸ”§ Fonctionnement du script

- Charge toutes les feuilles du fichier Excel (`sheet_name=None`)
- Ajoute un `station_id` fixe : **IICHTE19**
- Ajoute une colonne `date` correspondant au nom de la feuille
- Renomme les colonnes pour correspondre au schÃ©ma du pipeline
- ConcatÃ¨ne toutes les feuilles en un seul DataFrame
- Exporte le rÃ©sultat en JSON formatÃ©


```python
import pandas as pd

excel_file = "Weather Underground - Ichtegem, BE.xlsx"

# Charger toutes les feuilles
sheets = pd.read_excel(excel_file, sheet_name=None)

all_data = []

for sheet_name, df in sheets.items():

    # Ajouter station_id
    df["station_id"] = "IICHTE19"

    # Ajouter la colonne date = nom de la feuille
    df["date"] = sheet_name

    # Renommer les colonnes pour correspondre Ã  ton pipeline
    df = df.rename(columns={
        "Time": "time",
        "Temperature": "temperature",
        "Dew Point": "dew_point",
        "Humidity": "humidity",
        "Wind": "wind_direction",
        "Speed": "wind_speed",
        "Gust": "wind_gust",
        "Pressure": "pressure",
        "Precip. Rate.": "precip_rate",
        "Precip. Accum.": "precip_accum",
        "UV": "uv",
        "Solar": "solar"
    })

    # Ajouter au tableau global
    all_data.append(df)

# Fusionner toutes les feuilles
final_df = pd.concat(all_data, ignore_index=True)

# Export JSON
final_df.to_json(
    "ichtegem.json",
    orient="records",
    indent=4,
    force_ascii=False
)

print("Conversion terminÃ©e â†’ ichtegem.json")
```
